```python
#Importing the required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the dataset

df = sns.load_dataset('titanic')
```

```python
df1 = df.drop(['deck'], axis=1)
df1
```

```python
df1['age'].fillna(int(df1['age'].mean()), inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1.isnull().sum().sum()
```

```python
df1 = df1.dropna()
```

```python
df1.isnull().sum().sum()
```

```python
#Plotting a RugPlot

sns.rugplot(df1['fare'])
```

```python
#Plotting a SwarmPlot with hue as Survived and split as true

sns.swarmplot(x='sex', y='age', data=df1, hue='survived')
```

```python
#Combining a ViolinPlot and a SwarmPlot

sns.violinplot(x='sex', y='age', data=df1)
sns.swarmplot(x='sex', y='age', data=df1, color='black')
```
