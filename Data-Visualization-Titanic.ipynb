```python
#Importing the required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the dataset

df = sns.load_dataset('titanic')
```

```python
df1 = df.drop(['deck'], axis=1)
```

```python
df1['age'].fillna(int(df1['age'].mean()), inplace=True)
```

```python
df1 = df1.dropna()
```

```python
df1.shape
```

```python
#Plotting a DistPlot

sns.distplot(df1['fare'])
```

```python
#Plotting a JointPlot

sns.jointplot(x='age', y='fare', data=df1)
```

```python
#Plotting a BarPlot

sns.barplot(x='sex', y='age', data=df1)
```

```python
#Plotting a CountPlot

sns.countplot(x='sex', data=df1)
```

```python
#Plotting a BoxPlot

sns.boxplot(x='sex', y='age', data=df1)
```

```python
#Plotting a BoxPlot with hue as Survived

sns.boxplot(x='sex', y='age', data=df1, hue="survived")
```

```python
#Plotting a ViolinPlot

sns.violinplot(x='sex', y='age', data=df1)
```

```python
#Plotting a ViolinPlot with hue as Survived

sns.violinplot(x='sex', y='age', data=df1, hue='survived')
```

```python
#Plotting a BoxPlot with hue as Survived and split as true

sns.violinplot(x='sex', y='age', data=df1, hue='survived', split=True)
```

```python
#Plotting a StripPlot

sns.stripplot(x='sex', y='age', data=df1)
```

```python
#Plotting a StripPlot with hue as Survived and split as true

sns.stripplot(x='sex', y='age', data=df1, jitter=True, hue='survived')
```

```python
#Combining a ViolinPlot and a SwarmPlot

sns.violinplot(x='sex', y='age', data=df1)
sns.swarmplot(x='sex', y='age', data=df1, color='black')
```
